```python
import requests
import pandas as pd

#Endpoint da API
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)
df.head()
```

```python
# Apenas colunas relevantes
df_transformado = df[['id', 'name', 'email', 'address']].copy()

#Extraindo cidade do endereço
df_transformado["city"] = df_transformado["address"].apply(lambda x: x["city"])

#Removendo coluna desnecessária
df_transformado = df_transformado.drop(columns=["address"])

df_transformado
```

```python
df_transformado.to_csv("relatório_clientes_api.csv", index =False)
df_transformado
```
